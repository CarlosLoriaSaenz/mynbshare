```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(1, 5, 15)
Y = X * X
Z = 8 * np.log(X)
print(Z)
```

Ahora vamos a graficar tres funciones

```python
plt.plot(X, Y, label="x*x")
plt.plot(X, X, label="x")
plt.plot(X, Z, label="8*log(x)")
plt.title("Mi primera grafica")
plt.legend()
plt.show()
```
